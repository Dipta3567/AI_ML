Dataset : Practice Day (1).csv

https://github.com/phitronio/Python-for-ML/blob/main/Practice%20Day%20%281%29.csv

```python
import pandas as pd

url = "https://raw.githubusercontent.com/phitronio/Python-for-ML/main/Practice%20Day%20(1).csv"
df = pd.read_csv(url)

df.head()
```

# 1. Add New Column and Save File
You have a DataFrame with column Price.
 Create a new column Discounted_Price with (10% discounted price).
 Save the modified DataFrame as discounted_products.csv.


```python
# 10% discount -> price * .90

df["Discounted_Price"] = df["Price"]* 0.90

# save file
df.to_csv("discounted_products.csv", index=False)

df[["Price", "Discounted_Price"]].head()
```

# 2. Check Duplicates and Remove Them

Column: EmployeeID

## Count duplicates

```python
cnt= df["EmployeeID"].duplicated().sum()
print(cnt)
```

## Remove duplicates

```python
clr= df.drop_duplicates(subset="EmployeeID")
```

## Display cleaned DataFrame

```python
clr.head()
```

# 3. Handle Missing Values

Columns: Age, Salary

## Count missing values

```python
print(df[["Age", "Salary"]].isnull().sum())
```

## Fill missing Age with mean age

```python
df["Age"].fillna(df["Age"].mean())
```

## Drop rows where Salary is missing

```python
df = df.dropna(subset="Salary")
```

# 4. Apply Statistical Functions

Columns: Marks1, Marks2, Marks3

## Mean, Max for each subject

```python
mean_marks = df[["Marks1", "Marks2", "Marks3"]].mean()
max_marks = df[["Marks1", "Marks2", "Marks3"]].max()

print("Mean Marks:")
print(mean_marks)
print("\nMax Marks:")
print(max_marks)
```

## Compute total marks per student

```python
df["Total_Marks"]= df["Marks1"]+df["Marks2"]+df["Marks3"]
df.head()
```

## Standard deviation of total marks

```python
df["Total_Marks"].std()
```

# 5. Plots (Histogram, Scatter, Pie)

```python
import matplotlib.pyplot as plt
```

## Histogram of Age

```python
plt.hist(df["Age"])
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Scatter Plot (Age vs Salary)

```python
plt.scatter(df["Age"], df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

## Scatter Plot (Price vs Discounted Price)

```python
plt.scatter(df["Price"], df["Discounted_Price"])
plt.title("Price vs Discounted Price")
plt.xlabel("Price")
plt.ylabel("Discounted Price")
plt.show()
```
