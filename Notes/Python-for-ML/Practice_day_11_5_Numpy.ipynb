1. Create 1D, 2D, 3D arrays and print dimensions + shapes

```python
import numpy as np
# 1D
arr1 = np.array([1, 2, 3])
print(arr1.ndim, arr1.shape)

# 2D
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(arr2.ndim, arr2.shape)

# 3D
arr3 = np.array([
    [[1, 2], [3, 4]],
    [[5, 6], [7, 8]]
])
print(arr3.ndim, arr3.shape)
```

2)  Create:
an array of zeros of shape (3,3)


an array of ones of shape (2,5)


an array filled with 7


```python
zeros_arr = np.zeros((3, 3))
ones_arr = np.ones((2, 5))
filled_arr = np.full((3, 3), 7)

print(zeros_arr)
print(ones_arr)
print(filled_arr)
```

Create a 4×4 matrix from 0–15. Then , extract:
the first row


the last column


a 2×2 subarray from the center


```python
mat = np.arange(16).reshape(4, 4)
print(mat)

first_row = mat[0]
last_col = mat[:, -1]
center_2x2 = mat[1:3, 1:3]

print(first_row)
print(last_col)
print(center_2x2)
```

 Add a 1D array [1,2,3] to each row of a 3×3 matrix.


```python
matrix = np.arange(9).reshape(3, 3)
add_arr = np.array([1, 2, 3])

result = matrix + add_arr
print(result)
```

Given two arrays A and B, compute:
(A + B) / 2


A**2 + B**2


```python
A = np.array([1, 2, 3])
B = np.array([4, 5, 6])

part1 = (A + B) / 2
part2 = A**2 + B**2

print(part1)
print(part2)
```
