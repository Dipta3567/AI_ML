```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

# NumPy

### Ans to the Q1

```python
ages_arr= np.array(ages)
salaries_arr= np.array(salaries)

print("Ages: ", ages_arr)
print("Salaries: ", salaries_arr)
print("\nAges -> \n     dtype: ", ages_arr.dtype)
print("     ndim: ", ages_arr.ndim)
print("     shape: ", ages_arr.shape)
print("     size: ", ages_arr.size)

print("\nSalaries -> \n     dtype: ", salaries_arr.dtype)
print("     ndim: ", salaries_arr.ndim)
print("     shape: ", salaries_arr.shape)
print("     size: ", salaries_arr.size)
```

### Ans to the Q2

```python
high_sal= np.max(salaries_arr)
low_sal= np.min(salaries_arr)
avg_sal= np.mean(salaries_arr)
avg_age= np.mean(ages_arr)

print("Highest Salary: ", high_sal)
print("Lowest Salary: ", low_sal)
print("Average Salary: ", avg_sal)
print("Average Age: ", avg_age)
```

### Ans to the Q3

```python
age_above_30= ages_arr[ages_arr>30]
count_above_30 = age_above_30.size

print("Ages > 30:", age_above_30)
print("Number of employees > 30:", count_above_30)
```

### Ans to the Q4

```python
new_ages= ages_arr+5
print("Orginal: ", ages_arr)
print("New: ", new_ages)
```

### Ans to the Q5

```python
total= np.sum(salaries_arr)
diff= np.max(salaries_arr)-np.min(salaries_arr)

print("Total Salaries: ", total)
print("Difference between Max and Min Salary: ", diff)
```
