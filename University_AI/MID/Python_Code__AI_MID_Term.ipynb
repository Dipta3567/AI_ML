#### 3. Function Problem (Pet Greeting üê∂üê±)
#### Write a function pet_greet(pet_name, pet_type) that prints a greeting.
#### pet_greet("Tommy", "dog")
#### Hello Tommy the dog! üê∂ You are such a good pet!

```python
def pet_greet(name, pet_type):
    if pet_type == "dog":
        print(f'Hello {name} the {pet_type}! üê∂ You are such a good pet!')
    else:
        print(f'Hello {name} the {pet_type}! üò∫ You are such a good pet!')

Name = input("Enter the Name of the Pet = ")
Type = input("Enter the Type of the Pet = ")
Type = Type.lower()
pet_greet(Name, Type)
```

```python
#Scikit-learn example code:

from sklearn.linear_model import LinearRegression

# Step 1: Define the data
# X = input numbers, y = output numbers
X = [[1], [2], [3], [4], [5]]
y = [2, 4, 6, 8, 10]  # Pattern: y = 2 * x

# Step 2: Create the model
model = LinearRegression()

# Step 3: Train the model
model.fit(X, y)

# Step 4: Make predictions
prediction = model.predict([[6]])  # What comes after 5?
print("Prediction for input 6:", prediction[0])
```
