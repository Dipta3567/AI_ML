# BEST_FIRST_SEARCH ALGORITHM

```python
SuccList = {
    'A' : [['B',3], ['C', 2]],
    'B' : [['A',5], ['C',2], ['D',2], ['E',3]],
    'C' : [['A',5], ['B',3], ['F',2], ['G',4]],
    'D' : [['H',1], ['I',99]],
    'F' : [['J',99]],
    'G' : [['K',99], ['L',3]]
}
Start = 'A'
Goal = 'E'
closed = list()
SUCCESS = True
FAILURE = False
State = FAILURE

def MOVEGEN(N):
    New_list = list()
    if N in SuccList.keys():
        New_list = SuccList[N]
    return New_list

def GOALTEST(N):
    return N == Goal

def APPEND(L1, L2):
    return list(L1) + list(L2)

def SORT(L):
    L.sort(key = lambda x : x[1])
    return L

def BestFirstSearch():
    OPEN = [[Start, 5]]
    CLOSED = list()
    global State
    global Closed

    while (len(OPEN) != 0) and (State != SUCCESS):
        print ("--------------")
        N = OPEN[0]
        print("N = ", N)
        del OPEN[0]

        if GOALTEST(N[0]) == True:
            State = SUCCESS
            CLOSED = APPEND(CLOSED, [N])
            print("CLOSED = ", CLOSED)

        else:
            CLOSED = APPEND(CLOSED, [N])
            print("CLOSED = ", CLOSED)
            CHILD = MOVEGEN(N[0])
            print("CHILD = ", CHILD)

            for val in CLOSED:
                if val in CHILD:
                    CHILD.remove(val)

            for val in OPEN:
                if val in CHILD:
                    CHILD.remove(val)

            OPEN = APPEND(CHILD, OPEN)
            print("Unsorted OPEN = ", OPEN)

            SORT(OPEN)
            print("Sorted OPEN = ", OPEN)

    Closed = CLOSED
    return State


'''

result = BestFirstSearch()
print(Closed, result)

'''

result = BestFirstSearch()
print("\nVisited Nodes (CLOSED):", Closed)
print("Search Result: ", "SUCCESS" if result else "FAILURE")



```

# A* (A-Star) Algorithm

```python
Graph_nodes = {
    'A' : [['B', 6], ['F',3]],
    'B' : [['A',6], ['C', 3], ['D', 2]],
    'C' : [['B',3], ['D',1], ['E',5]],
    'D' : [['B',2], ['C', 1], ['E', 8]],
    'E' : [['D',8], ['C', 5], ['I', 5], ['J', 5]],
    'F' : [['G',1], ['H', 7]],
    'G' : [['F',1], ['I', 3]],
    'H' : [['F',7], ['I', 2]],
    'I' : [['E',5], ['G', 3], ['H', 2], ['J', 3]],
    'J' : [['E',5], ['I', 3]],
}

H_dist = {
    'A': 10, 'B': 8, 'C': 5, 'D': 7, 'E': 3,
    'F': 6, 'G': 5, 'H': 3, 'I': 1, 'J': 0
}

start = 'A'
goal = 'J'

# ----- A* Search Implementation -----
def A_star_search(start, goal):
    OPEN = [[start, 0, H_dist[start], 0 + H_dist[start]]]  # [node, g, h, f]
    CLOSED = []

    print("Initial OPEN:", OPEN)
    print("---------------------------")

    while OPEN:
        # Step 1: Sort OPEN based on f = g + h
        OPEN.sort(key=lambda x: x[3])

        # Step 2: Pick the node with smallest f
        current = OPEN.pop(0)
        node, g, h, f = current
        CLOSED.append(current)

        print(f"Selected Node: {node} | g={g}, h={h}, f={f}")
        print("CLOSED:", [x[0] for x in CLOSED])

        # Step 3: Goal test
        if node == goal:
            print("\nâœ… Goal Reached!")
            print("Final Path:", [x[0] for x in CLOSED])
            print("Total Cost:", g)
            break

        # Step 4: Expand children
        children = Graph_nodes.get(node, [])
        temp_list = []

        for (child, cost) in children:
            g_new = g + cost
            h_new = H_dist[child]
            f_new = g_new + h_new

            # Check if already visited
            if any(child == n[0] for n in CLOSED):
                continue

            # If already in OPEN with higher cost, replace
            in_open = [n for n in OPEN if n[0] == child]
            if in_open:
                if f_new < in_open[0][3]:
                    OPEN.remove(in_open[0])
                    OPEN.append([child, g_new, h_new, f_new])
            else:
                OPEN.append([child, g_new, h_new, f_new])
            temp_list.append([child, g_new, h_new, f_new])

        print("Generated Children:", temp_list)
        print("OPEN:", [[n[0], n[3]] for n in OPEN])
        print("---------------------------")

    return CLOSED


# ----- Run the Search -----
result = A_star_search(start, goal)
```

# Hill Climbing

```python
import random

class EightQueens:

    def __init__(self):
        #Initialize the board with 8 Queen in each row, placed them randomly in of the columns.
        self.board = [random.randint(0, 7) for _ in range(8)]
    '''

    def __init__(self, initial_board):
        """
        initial_board: Optional list of 8 integers (0-7) representing the column of queen in each row.
        If not provided, generate random board.
        """
        if initial_board:
            self.board = initial_board
        else:
            self.board = [random.randint(0, 7) for _ in range(8)]
    '''

    def print_board(self):
        for row in range(8):
            line = ""
            for col in range(8):
                if self.board[row] == col:
                    line += 'Q '
                else:
                    line += '. '
            print(line)
        print('\n')

    def calculate_conf(self, board):
        #calculating the number of conflicts (Queens Attacking each other) for a given board state.
        conflicts = 0
        for row1 in range(8):
            for row2 in range(row1+1, 8):
                # Check if same Column
                if board[row1] == board[row2]:
                    conflicts += 1
                # Check if same Diagonal
                if abs(board[row1] - board[row2]) == abs(row1 - row2):
                    conflicts += 1
        return conflicts

    def Hill_Climb(self):
        #Perform simple hill climbing until we reach a solution or local maxima

        current_conflicts = self.calculate_conf(self.board)

        while True:
            # Looking for better neighbour if we find....
            found_better = False

            #Try to moving each queen ( in each row) to a different column for better replacement
            for row in range(8):
                real_col = self.board[row]
                for col in range(8):
                    if col != real_col: #Moving different column
                        new_board = self.board[:] #Copying the current board
                        new_board[row] = col #Move the queen to the new column
                        new_conflicts = self.calculate_conf(new_board)

                        # If we find a better board
                        if new_conflicts < current_conflicts:
                            self.board = new_board
                            current_conflicts = new_conflicts
                            found_better = True
                            break # Move on without checking further neighbors

                    if found_better:
                        break # Move for the new iteration wih the new board

                if found_better == False:
                    break # If there any better neighbour exist

            return current_conflicts == 0

#Examples Usage:
'''
manual_board = [3, 5, 7, 1, 6, 0, 2, 4]  # Q1-Q8 columns in row 0-7
solve = EightQueens(manual_board) #Create an 8 queens solver object
'''
solve = EightQueens()
print("Initial Board: ")
solve.print_board() #Print the initial random board

if solve.Hill_Climb(): #The Algorithm....
    print("Solution Found: ")
else:
    print("Local Maximum reached, no solution found: ")
solve.print_board() #Print the final board (solution or local maximum)


```

# **Uniform Cost Search Algorithm**

```python
from queue import PriorityQueue
class Graph:
    def __init__(self, succ_list, directed = False):
        """Initialize Graph directly from a given successor list (dictionary)"""
        self.graph = succ_list
        self.directed = directed

    def ucs(self, start, goal):
        '''
        Uniform Cost Search (UCS) with OPEN & CLOSED List Display
        '''
        open_list = PriorityQueue() #(Cost, Node, path)
        open_list.put((0, start, [start])) # As PriorityQueue so we used put function(like append)
        closed_list = []

        print("Step-by-Step UCS Execution : \n")
        step = 1  # Step counter

        while not open_list.empty():
            cost, node, path = open_list.get() #As PriorityQueue so we used get function(like pop)

            if node in closed_list:
                continue

            closed_list.append(node)

            #Display Current State
            print(f"------- Step : {step} -------")
            print(f'Current Node : {node}')
            print(f'   Path So Far : {path}')
            print(f'   Total Cost : {cost}')
            print(f'   CLOSED : {closed_list}')

            #Check goal node
            if node == goal:
                print()
                print("-----------------")
                print("Goal Reached!!!")
                print(f'Optimal Path : {' -> '.join(path)}')
                print(f'Total Cost : {cost}')
                return

            #Explore Neighbours
            for neighbour, weight in self.graph.get(node, []):
                if neighbour not in closed_list:
                    new_cost = cost + weight
                    new_path = path + [neighbour]
                    open_list.put((new_cost, neighbour, new_path))

            #Show OPEN List for Clarity
            open_temp = list(open_list.queue)
            print(f' OPEN : {[ (c, n) for c, n, _ in open_temp ]}')
            print()
            step += 1

        print('Goal not Found!!!')


# ðŸ”¹ Declare the graph directly as a successor list
SuccList = {
    'S': [['A', 2], ['D', 5]],
    'A': [['S', 2], ['D', 2], ['B', 1]],
    'B': [['A', 1], ['C', 4], ['E', 5]],
    'D': [['S', 5], ['A', 2], ['E', 2]],
    'E': [['D', 2], ['B', 5], ['F', 4]],
    'F': [['E', 4], ['G', 3]]
}
'''
SuccList = {
    'V1': [['V2', 9], ['V3', 4]],
    'V2': [['V1', 9], ['V3', 2], ['V5', 3], ['V4', 7]],
    'V3': [['V1', 4], ['V2', 2], ['V4', 1], ['V5', 6]],
    'V4': [['V2', 7], ['V3', 1], ['V5', 4], ['V6', 8]],
    'V5': [['V2', 3], ['V3', 6], ['V4', 4], ['V6', 2]],
    'V6': [['V4', 8], ['V5', 2]]
    # Start = V1, Goal = V6
}

'''

# ðŸ”¹ Run UCS
g = Graph(SuccList, directed = False)
g.ucs('S', 'G')
```
