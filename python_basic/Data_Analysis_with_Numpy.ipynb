 | Option | Meaning                                           | Flattening order                                           |
 | :----- | :------------------------------------------------ | :--------------------------------------------------------- |
 | `'C'`  | **C order** (row-major, default)                  | Row by row (left  right, top  bottom)                    |
 | `'F'`  | **Fortran order** (column-major)                  | Column by column (top  bottom, left  right)              |
 | `'A'`  | **Any order** (depends on memory layout of array) | Uses row-major if array is C-contiguous, else column-major |
 | `'K'`  | **Keep order**                                    | Follows the actual memory layout (most raw form)         |

```python
import numpy as np

def matrix_dot_vector(a: list[list[int|float]], b: list[int|float]) -> list[int|float]:
	# Return a list where each element is the dot product of a row of 'a' with 'b'.
	# If the number of columns in 'a' does not match the length of 'b', return -1.

    a = np.array(a)
    b=np.array(b)

    shape_a = a.shape
    if shape_a[1]!=b.size:
        return -1

    dot = np.dot(a,b).tolist()
    return dot

```
